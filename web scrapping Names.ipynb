```python
import csv
from requests import get
from bs4 import BeautifulSoup
```

```python
li = []
for j in range(1, 105):
    url = "https://hamariweb.com/names/muslim/girl/page-{}".format(j)
    response = get(url)
    #print(response)
    material  =  BeautifulSoup(response.text,'html.parser')
   # print(material)
    name_col = material.find_all('div',class_="NewPinkBox white text-center")
    fm = []
    for i in name_col:
        for j in i.findChildren("a",class_= "white UrduLinkBold"):
            fm.append(j.text)
    with open("C:/Users/M.Bytes/Desktop/female.txt",'a',encoding="utf-8") as f:
        for n in fm:
            f.write(n)
    li.append(fm)
```

```python
len(li)
```

```python
for i in range(len(li)):
    for j in li[i]:
        print(j)
```
